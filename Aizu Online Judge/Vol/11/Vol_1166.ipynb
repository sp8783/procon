```python
from collections import deque

dxy = [(1,0),(0,1),(0,-1),(-1,0)]

def bfs(w, h, ver, hor):
    dist = [[0]*w for _ in range(h)]
    dist[0][0] = 1
    q = deque()
    q.append((0,0))
    while q:
        y, x = q.popleft()
        if y == h-1 and x == w-1:
            break
        else:
            for dx, dy in dxy:
                if (dx==-1 and ver[y][x]==0) or (dx==1 and ver[y][x+1] == 0) or (dy==-1 and hor[y][x]==0) or (dy==1 and hor[y+1][x]==0): # 通行可能か
                    if dist[y+dy][x+dx] == 0:
                        q.append((y+dy, x+dx))
                        dist[y+dy][x+dx] = dist[y][x] + 1
    return dist[h-1][w-1]

ans = []
while True:
    w, h = map(int, input().split())
    if w == 0 and h == 0:
        break
    ver = []
    hor = []
    hor.append([1]*w)
    for i in range(2*h-1):
        bar = list(map(int, input().split()))
        if i%2 == 0:
            ver.append([1] + bar + [1])
        else:
            hor.append(bar)
    hor.append([1]*(w+1))
    ans.append(bfs(w, h, ver, hor))

for a in ans:
    print(a)
```

```python
from collections import deque

while True:
    w, h = map(int, input().split())
    
    # 終了条件
    if w == 0 and h == 0:
        break
        
    # 「要素が空白であるマス -> 移動先候補」,「要素が 0 or 1 である -> 壁の有無」として迷路をinput
    maze = []
    for i in range(2*h-1):
        lst = list(input())
        if i%2 == 0:
            lst = [" "] + lst + [" "]
        maze.append(lst)

    q = deque()
    q.append([0,0])
    dxy = [(1,0),(0,1),(0,-1),(-1,0)]
    
    # 各移動先候補(w*h)の最短経路を格納
    ans = [[-1]*w for _ in range(h)]     
    ans[0][0] = 1
    
    # ゴール到達フラグを初期化
    flag = False     

    # 幅優先探索
    while q:
        x, y = q.popleft()
        
        # ゴールに到達した場合
        if x == 2*w - 2 and y == 2*h - 2:
            flag = True
            break
            
        for dx, dy in dxy:
            # 壁が存在し得る座標群
            ndx = x + dx
            ndy = y + dy
            # 移動先となり得る座標群
            nx = x + 2 * dx
            ny = y + 2 * dy
            
            # 移動先が存在しない場合
            if ndx < 0 or 2*w - 2 < ndx or ndy < 0 or 2*h - 2 < ndy:
                continue
            # 移動先に壁がある場合
            if maze[ndy][ndx] == "1":
                continue
            # 既に探索済みである場合
            if ans[ny//2][nx//2] != -1:
                continue
            
            ans[ny//2][nx//2] = ans[y//2][x//2] + 1
            q.append([nx, ny])
    
    if flag:
        print(ans[h-1][w-1])
    else:
        print(0)
```
